```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/processed/ncbi_studies_bioprojects.txt')
```

```python
with open(path,'r') as file:
    lista = file.read()
access_codes = lista.split('\n')
```

```python
access_codes
```

```python
import requests

def request_data(project_ids):
    lista = []
    for id in project_ids: 
        url = f'https://www.ebi.ac.uk/metagenomics/api/v1/studies/{id}/downloads'
        response = requests.get(url)
        data = response.json()
        lista.append(data)
    return lista

request_data(access_codes)
```
